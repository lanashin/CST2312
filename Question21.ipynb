```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
!curl "https://raw.githubusercontent.com/ProfessorPatrickSlatraigh/data/main/MinWageData.csv" -o MinWageData.csv
```

```python
data = pd.read_csv("MinWageData.csv")
```

```python
data.head()
```

```python
data.describe
```

```python
#Identify the state with the highest mean for minimum wage over the years in the dataset.

(
      data.groupby('State')
  .agg(Average_Wage = ('StateMinimumWage', 'mean'), 
    )
  .head(50)
  .sort_values("Average_Wage", ascending=False)
)
```

```python
state_year = pd.DataFrame()
state_year = data.loc[data['State'].isin([('Alaska')])]
print(state_year.head)
```

```python
state_year.mean()
```

```python
#check if DC will show the same data

state_year_dc = pd.DataFrame()
state_year_dc = data.loc[data['State'].isin([('District of Columbia')])]
print(state_year_dc.head)
```

```python
state_year_dc.mean()
```

```python
#Produce a line plot of the minimum wage for the state you identified as having the highest mean value for minimum wage over the years in the dataset

fig = plt.figure(figsize=(10, 10))
ax = fig.gca()                   
state_year.plot(x = 'Year', y = 'StateMinimumWage', ax=ax, color = "orange") 

ax.set_title('Alaska Minimum Wage') 
ax.set_xlabel('Year') 
ax.set_ylabel('Wage')
```
